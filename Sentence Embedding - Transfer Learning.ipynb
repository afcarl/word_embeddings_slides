```python
import multiprocessing
from gensim.corpora.wikicorpus import WikiCorpus
from gensim.models.word2vec import LineSentence
from gensim.models.word2vec import Word2Vec
import numpy as np
import gensim
```

```python
sentences = gensim.models.word2vec.Text8Corpus('text8')
```

```python
model = gensim.models.Word2Vec(sentences, min_count=10, size=64, workers=4)
```

```python
sentences_list = list(sentences)
```

```python
vector_sentences = []
for sent in sentences_list:
    vec = np.zeros(64)
    for word in sent:
        if (word in model):
            vec += model[word]
    vec /= len(sent)
    vector_sentences.append(vec)
```

```python
import math

def dot_product(v1, v2):
    return sum(map(lambda x: x[0] * x[1], zip(v1, v2)))

def cosine_measure(v1, v2):
    prod = dot_product(v1, v2)
    len1 = math.sqrt(dot_product(v1, v1))
    len2 = math.sqrt(dot_product(v2, v2))
    return prod / (len1 * len2)
```

```python
ind = -1
max_sim = -1
vec1 = vector_sentences[0]
for i in range(1, len(vector_sentences)):
    vec2 = vector_sentences[i]
    sim = cosine_measure(vec1, vec2)
    if (sim > max_sim):
        max_sim = sim
        ind = i
```

```python
from gensim.models.keyedvectors import KeyedVectors
model = KeyedVectors.load_word2vec_format('word2vec_glove.txt', binary=False)
```

```python
sentences = ["code is beautiful", "computers are good", "nice code", "he went to chicago", "he stays at chicago", "he came here for code", "he was coding"]
```

```python
vector_sentences = []
for sent in sentences:
    vec = np.zeros(50)
    for word in sent.split():
        if (word in model):
            vec += model[word]
    vec /= len(sent)
    vector_sentences.append(vec) 
```

```python
def most_similar_to_ind(inp_ind):
    ind = -1
    max_sim = -1
    vec1 = vector_sentences[inp_ind]
    for i in range(len(vector_sentences)):
        if (i == inp_ind):
            continue
        vec2 = vector_sentences[i]
        sim = cosine_measure(vec1, vec2)
        if (sim > max_sim):
            max_sim = sim
            ind = i
    return ind
```

```python
for ind in range(len(sentences)):
    print (sentences[ind], "- is most similar to -",sentences[most_similar_to_ind(ind)])
```
